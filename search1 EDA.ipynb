```python
# 시각화 준비
import matplotlib
import matplotlib.pyplot as plt
# Jupyter Notebook 내부에 그래프를 출력하도록 설정
%matplotlib inline
import seaborn as sns
```

```python
# 코드 1-8. 변수를 막대 그래프로 시각화하기

cols = ['SEARCH_CNT']
for col in cols:

    # 그래프 크기 (figsize) 설정
    f, ax = plt.subplots(figsize=(20, 15))
    # seaborn을 사용한 막대그래프 생성
    sns.countplot(x=col, data=sea, alpha=1)
    # show() 함수를 통해 시각화
    plt.show()
```
